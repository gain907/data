```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/순천대학교/data/titanic.csv')
df
```

```python
# 데이터프레임 결측값 유무 확인
df.info()
```

```python
df.describe()
```

- 데이터 전처리
- Age컬럼 결측지 평균값으로 대치
- Embarked 결측지 최빈값으로 대치
- Cabin 컬럼은 결측지 많아서 제거 할 거임

```python
# Age 컬럼 결측값 평균
d_mean = df['Age'].mean()
df['Age'].fillna(d_mean, inplace= True)
# Embarked 결측지 최빈값으로 대치
d_mode = df['Embarked'].mode()[0]
print(d_mode)
df['Embarked'].fillna(d_mode, inplace=True)
```

```python
d_mode = df['Embarked'].mode(2)
print(d_mode)
```

```python
# Sex 컬럼, Embarked 컬럼 labelEncoder 처리 할 거임
from sklearn.preprocessing import LabelEncoder
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])


# Sex 컬럼, Embarked 컬럼 labelEncoder 처리 할 거임
from sklearn.preprocessing import LabelEncoder
df['Embarked'] = LabelEncoder().fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
# SiSP, Parch의 값을 더해서 FamilySize 컬럼생성

df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
df
```

```python
# 분석 데이터 셋 준비
# 문제 답지 분류

X = df[['Pclass','Sex','Age','Fare','Embarked','FamilySize']]
y = df['Survived']
```

```python
X
```

```python
y
```

```python
# 데이터 8:2 비율로 분할
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 데이터 분석 수행
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train,y_train)
```

```python
pred = dt.predict(X_test)
print(pred)
```

```python
# 모델 성능 - 정확도 측정

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,pred)
print(acc)
```

# KNN 알고리즘으로 iris데이터 셋 학습해 보기
- 거리기반 알고리즘이기 때문에 값들을 표준화 시켜주고 모델을 실행시켜야 한다
- 거리기반의 연산으로 , 숫자로 구분된 속성에 우수한 성능을 보임
-
- 데이터 정보
 - sepal_length : 꽃받침 길이
 - sepal_width  : 꽃받침 넓이
 - petal_length : 꽃잎의 길이
 - petal_width  : 꽃입의 넓이

```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/순천대학교/data/iris.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# 데이터 전처리
# 요번에는 min-max로 데이터를 정규화시킴(0~1사이로)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['sepal_length']] = scaler.fit_transform(df[['sepal_length']])
df[['sepal_width']] = scaler.fit_transform(df[['sepal_width']])
df[['petal_length']] = scaler.fit_transform(df[['petal_length']])
df[['petal_width']] = scaler.fit_transform(df[['petal_width']])
```

```python
df.head()
```

```python
# 문제 답지 분리
X = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df['species']

# 분활 8:2
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 모델
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
print(pred)
```

```python
# 성능평가
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,pred)
print(acc)
```

### 추가 작업 해보기[직접]
- 타이타닉 데이터셋의 생존 여부 분류 문제 해결
- 기존모델은 의삭셜정나무 모델
- KNN 모델로 바꿔서 비교해 보기

### 변경 전

```python
# 변경전
from sklearn.tree import DecisionTreeClassifier
# 데이터 분석 수행
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train,y_train)

# 데이터 예측 수행
pred = dt.predict(X_test)
print(pred)
```

### 변경 후

```python
from sklearn.neighbors import KNeighborsClassifier
# 데이터 분석 수행
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)

# 데이터 예측 수행
pred = knn.predict(X_test)
print(pred)
```

### 오차행렬을 이용한 성능평가

```python
# 모델 성능 평가 - Confusion_Maxtrix
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test,pred)
print(mat)
```

### 평가지표 계산
 - precision,recall,f1-score,support 구하기

```python
# 모델 성능 평가 - 평가지표 계산
from sklearn.metrics import classification_report
rpt = classification_report(y_test,pred)
print(rpt)
```

### 판다스 통계함수(가장 빈번히 사용되는 함수)
- 최대값 .max
- 최소값 .min
- 평균값 .mean
- 중앙값 .median
- 최빈값 .mode
- 합계 .sum
- 데이터 수 .count

```python
print('최대값')
print('f1',df['f1'].max())
print('f5',df['f5'].max())
print('age',df['age'].max())
```

```python
print('최소값')
print('f1',df['f1'].min())
print('f5',df['f5'].min())
print('age',df['age'].min())
```

```python
print('평균값')
print('f1',df['f1'].mean())
print('f5',df['f5'].mean())
print('age',df['age'].mean())
```

```python
print('중앙값')
print('f1',df['f1'].median())
print('f5',df['f5'].median())
print('age',df['age'].median())
```

```python
print('최빈값')
print('f1',df['f1'].mode())
print('f5',df['f5'].mode())
print('age',df['age'].mode())
```

- 분위수 .quantile
- 분산 .var
- 표준편차 .std
- 왜도: .skew 자료의 대칭성 오른쪽 꼬리 길면 + , 반대면 -
- 첨도: .kurt 뾰족함에 대한 정보를 제공

```python
print('분위수')
print('f1',df['f1'].quantile(.25))
print('f5',df['f5'].quantile(.75))
print('\n') # 엔터

print('분산')
print('f1',df['f1'].var())
print('f5',df['f5'].var())
print('\n') # 엔터

print('표준편차')
print('f1',df['f1'].std())
print('f5',df['f5'].std())
print('\n') # 엔터

print('왜도')
print('f1',df['f1'].skew())
print('f5',df['f5'].skew())
print('\n') # 엔터

print('첨도')
print('f1',df['f1'].kurt())
print('f5',df['f5'].kurt())
print('\n') # 엔터
```

### 판다스 통계함수
- 누적합 .cumsum
- 누적곱 .cumprod
- 누적최대값 .cummax
- 누적최소값 .cummin


```python
print('누적합')
# [:1] or [0:1] 첫번째 만 뽑아주세요
print('f1',df['f1'].cumsum()[:1])
print('f5',df['f5'].cumsum()[:1])
print('\n') # 엔터
print('누적곱')
# [:1] or [0:1] 첫번째 만 뽑아주세요
print('f1',df['f1'].cumprod()[:1])
print('f5',df['f5'].cumprod()[:1])
print('\n') # 엔터
print('누적 최대값')
# [:1] or [0:1] 첫번째 만 뽑아주세요
print('f1',df['f1'].cummax()[:1])
print('f5',df['f5'].cummax()[:1])
print('\n') # 엔터
print('누적 최소값')
# [:1] or [0:1] 첫번째 만 뽑아주세요
print('f1',df['f1'].cummin()[:1])
print('f5',df['f5'].cummin()[:1])
print('\n') # 엔터
```

### 판다스 통계함수

- 평균의 표준오차 .sem
- 평균 절대편차 .mad
- 절대값 .abs
- 곱 .prod

```python
print('평균의 표준 편차')
print('f1',df['f1'].sem())
print('f5',df['f5'].sem())
print('\n') # 엔터
print('평균 절대편차')
print('f1',df['f1'].mad())
print('f5',df['f5'].mad())
print('\n') # 엔터
print('절대값')
print('f1',df['f1'].abs())
print('f5',df['f5'].abs())
print('\n') # 엔터
print('곱')
print('f1',df['f1'].prod())
print('f5',df['f5'].prod())
```
